#Your First Visualization — Exploring Restaurant Tips

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
tips = sns.load_dataset("tips")

# Group by day and calculate total tips
tips_per_day = tips.groupby("day")["tip"].sum()

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(tips_per_day.index, tips_per_day.values, color="skyblue")

# Add labels and title
plt.title("Total Tips Collected by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Tip Amount")

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
tips = sns.load_dataset("tips")

# Create histogram with KDE overlay
plt.figure(figsize=(8, 5))
sns.histplot(tips["tip"], bins=20, kde=True, color="skyblue")

# Add title
plt.title("Distribution of Tip Amounts")

# Show chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
tips = sns.load_dataset("tips")

# Create scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", palette="Set2")

# Add title
plt.title("Total Bill vs Tip Amount")

# Show chart
plt.show()
```
